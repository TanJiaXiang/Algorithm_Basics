Matplotlib绘图，主要是基于`matplotlib.pyplot`方法。

```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
plt.plot([1,2,3,4])
plt.show()
```

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
```

颜色字符串与线型字符串连接在一起

```python
#“b-”
plt.plot([1, 2, 3, 4], [1, 4, 9, 16],"ro")
plt.axis([0, 6, 0, 20]) #[xmin，xmax，ymin，ymax]
```

```python
import numpy as np
t = np.arange(0., 5., 0.2)
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```

```python
#setp()
lines = plt.plot(t, t,'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.setp(lines,linewidth=5.0)
plt.show()
```
